# ETL Step 1 — Extract

Defines `extract(filepath)` — prints the CSV shape.

```python
from pathlib import Path
import csv

def extract(filepath):
    csv_path = Path(filepath)
    if not csv_path.exists():
        raise FileNotFoundError(f"Could not find CSV: {csv_path}")

    with csv_path.open(newline="", encoding="utf-8") as handle:
        reader = csv.reader(handle)
        header = next(reader)
        row_count = sum(1 for _ in reader)

    print("Shape:", (row_count, len(header)))
```

## Run extract on Lightcast dataset

```python
for candidate in [
    "lightcast_job_postings.csv",
    "../lightcast_job_postings.csv",
    "data/raw/lightcast_2024.csv",
    "../data/raw/lightcast_2024.csv",
]:
    if Path(candidate).exists():
        extract(candidate)
        break
else:
    raise FileNotFoundError("Could not find the Lightcast CSV in any expected location.")
```
