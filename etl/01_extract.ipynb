# ETL Step 1 — Extract

Defines `extract(filepath)` — loads a CSV, prints a full inspection report, and returns a DataFrame.

```python
import pandas as pd

def extract(filepath):
    df = pd.read_csv(filepath, low_memory=False)

    pd.set_option('display.max_columns', None)

    print("Shape:", df.shape)
    print("\nColumns:", df.columns.tolist())
    print("\nDtypes:\n", df.dtypes)
    print("\nHead:")
    display(df.head(5))
    print("\nTail:")
    display(df.tail(5))
    print("\nMissing values:\n", df.isnull().sum())
    print("\nBasic stats:")
    display(df.describe())

    return df
```

## Run extract on Lightcast dataset

```python
df = extract("../data/raw/lightcast_2024.csv")
```
