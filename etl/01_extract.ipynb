# ETL Step 1 — Extract

Defines `extract(filepath)` — prints the CSV shape.

```python
import pandas as pd

def extract(filepath, sample=None):
    if sample:
        df = pd.read_csv(filepath, nrows=sample, low_memory=False)
    else:
        chunks = []
        for chunk in pd.read_csv(filepath, chunksize=500_000, low_memory=False):
            chunks.append(chunk)
        df = pd.concat(chunks, ignore_index=True)

    pd.set_option('display.max_columns', None)
    print("Shape:", df.shape)
    print("\nColumns:", df.columns.tolist())
    print("\nDtypes:\n", df.dtypes)
    print("\nHead:")
    display(df.head(5))
    print("\nTail:")
    display(df.tail(5))
    print("\nMissing values:\n", df.isnull().sum())
    print("\nBasic stats:")
    display(df.describe())

    return df
```

## Run extract on Lightcast dataset

```python
df = extract("../lightcast_job_postings.csv", sample=10000)
```
